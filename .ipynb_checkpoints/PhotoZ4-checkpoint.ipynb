```python
import pandas as pd
import numpy as np
```

```python
template_data = pd.read_csv('template_dataset_150_steps_from0to1_approx45000lines.csv')
template_data.head(n=3)
```

```python
measured_data = pd.read_csv('SDSSDR7-10-percent-err-limit-25000-line.csv')
measured_data['redshift'] = measured_data['z'] # rename column z to redshift (name mismatch)
measured_data['r'] = measured_data['m_r']
measured_data[['r', 'ug', 'gr','ri', 'iz', 'redshift' ]].head(n=3)
```

#### In order to be able to use later on de column names for the Beck-method

```python
template_data['ug'] = template_data['u'] - template_data['g']
template_data['gr'] = template_data['g'] - template_data['r']
template_data['ri'] = template_data['r'] - template_data['i']
template_data['iz'] = template_data['i'] - template_data['z']
template_data[['r', 'ug', 'gr','ri', 'iz', 'redshift' ]].head(n=3)
```

### Representation of the function that will do the template selection based on real world, measured data.

```python
from sklearn.neighbors import NearestNeighbors

def find_nearest_template_in_measured_data(template_data, measured_data, indices):
    
    # NN algorithm
    neigh = NearestNeighbors(n_neighbors=1, algorithm='kd_tree', metric='l2')
    neigh.fit(template_data[indices].values)
    
    inds = neigh.kneighbors(measured_data[indices].values, return_distance=True)
    
    order = np.argsort(inds[0].flatten())
    
    N = 5000
    
    values = template_data[indices].values[inds[1].flatten()[order][:N], :].reshape(N, 5)
    
    df = pd.DataFrame(values)
    df.columns = indices
    
    return df
```

```python
df = find_nearest_template_in_measured_data(template_data, measured_data, ['ug', 'gr', 'ri', 'iz', 'redshift'])
```
